```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('diabetes.csv')


plt.figure(figsize=(8, 6))


diabetic = df[df['Outcome'] == 1]
non_diabetic = df[df['Outcome'] == 0]

plt.scatter(non_diabetic['Glucose'], non_diabetic['BMI'], color='blue', label='Non-Diabetic (0)', alpha=0.6)
plt.scatter(diabetic['Glucose'], diabetic['BMI'], color='red', label='Diabetic (1)', alpha=0.6)


plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.title('Scatter Plot of Glucose vs BMI (Colored by Outcome)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

cnf_matrix = confusion_matrix(y_test, y_pred)


class_names = ['Class A', 'Class B']  


df_cnf = pd.DataFrame(cnf_matrix, index=class_names, columns=class_names)


plt.figure(figsize=(8, 6))
sns.heatmap(df_cnf, annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```
