```python
data = pd.read_csv(r"/Users/krush/OneDrive/Desktop/used_cars_data.csv")

data = data.dropna(subset=['Year', 'Kilometers_Driven'])

data['Year'] = data['Year'].astype(int)
data['Kilometers_Driven'] = data['Kilometers_Driven'].astype(int)

bins = list(range(data['Year'].min(), data['Year'].max() + 3,3 ))

plt.hist(data['Year'], bins=bins, weights=data['Kilometers_Driven'], color='yellow',edgecolor='black')

plt.xlabel('Year (5-Year Bins)')
plt.ylabel('Total Kilometers Driven')
plt.title('Histogram of Kilometers Driven by Car Year')
plt.xticks(bins, rotation=45)
plt.tight_layout()

plt.show()

```
